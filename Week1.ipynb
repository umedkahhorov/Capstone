```python
import wikipedia as wp
import pandas as pd
```

```python
#html = wp.page('List_of_postal_codes_of_Canada:_M').html().encode('UTF-8')
#df = pd.read_html(html,header=0)[0]
url = 'http://zims-en.kiwix.campusafrica.gos.orange.com/wikipedia_en_all_nopic/A/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url,header=None)[0]
df.to_csv('List_of_postal_codes_of_Canada.csv')
df.head(2)
```

```python
coors = pd.read_csv('Geospatial_Coordinates.csv')
coors = coors.rename(columns={'Postal Code':'Postcode'})
coors.head(2)
```

```python
data = pd.merge(df,coors,on='Postcode')
data.head(2)
```

```python
import folium
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, TO'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_to = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(data.Latitude, data.Longitude, data.Borough, data.Neighbourhood):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_to)  
    
map_to
```
